# Singly LinkedLIst

```python
class Node:
    def __init__(self,data,next=None):
        self.data=data
        self.next=next 
# self basically refers to that object for which we going to perform operation/method   
class linkedlist:
    
    def __init__(self):
        self.head=None
# head stick with object say we create object dar or linked then head for that is saved because of self, 
# self.head is head for that object if self.head for dar changes then if will reflect in other instance of dar object too 
# similarly if we want to previous current point and want to save its position for next time or next object call
# then put it in constructor(__init__) along with head, self.current=None and now onward in methods use it as self.current
    def append(self,data):
        node=Node(data)
#         current= Node()
        if(self.head==None):
            self.head=node
            return
        else:
            current = self.head
            while(current.next!=None):
                current=current.next
            current.next=node
            return
        
    def display(self):
        if(self.head==None):
            print("no node")
            return
        else:
#             current= Node()
            current = self.head
            while(current!=None):
                print(current.data,end="->")
                current=current.next
            return 
        
    def pop(self):
        if(self.head==None):
            print("no element")
        elif(self.head.next==None):
            print('\n',self.head.data)
            data=self.head.data
            self.head=None
            return data
        else:
            current=self.head
            prev=self.head
            while(current.next!=None):
                prev=current
                current=current.next
            print('\n',current.data)
            data=current.data
            prev.next=None
            return data
            
dar=linkedlist()
dar.append(10)
dar.append(30)
dar.append(40)
dar.append(50)

dar.display()
dar.pop()
dar.pop()
dar.display()
```

# Doubly LinkedList

```python
class Node:
    def __init__(self):
        self.prev=None
        self.data=None
        self.next=None
        
class doublyLinkedList:
    
    def __init__(self):
        self.head=None
        
    def append(self,data):
        node=Node()
        current=self.head
        if(self.head==None):
            self.head=node
            node.data=data
            current=self.head
            return
        elif(self.head.next==None):
            node.prev=self.head
            self.head.next=node
            node.data=data
            return
        else:
            while(current.next!=None):
                current=current.next
            node.prev=current
            current.next=node
            node.data=data
            return
        
    def display(self):
        print("\n")
        if(self.head==None):
            print("no node")
            return
        else:
            current = self.head
            while(current!=None):
                print(current.data,end="->")
                current=current.next
            return 
    def pop(self):
        current=self.head
        while(current.next.next!=None):
            current=current.next
        print("\n",current.next.data)
        data=current.next.data
        current.next.prev=None
        current.next=None
        return data
    
    def insertAfter(self,data,newdata):
        current=self.head
        while(current.data!=data):
            current=current.next
            if(current.next==None):
                print("Node with data %d not found, unable to append node"%data)
                return
        node=Node()
        node.prev=current
        node.data=newdata
        node.next=current.next
        current.next.prev=node
        current.next=node
        return
            
link=doublyLinkedList()

link.append(10)
link.append(20)
link.append(30)
link.append(40)
link.append(50)
link.append(60)
print("after append:")
link.display()

link.pop()
# link.pop()
print("After pop:")
link.display()

link.insertAfter(40,80)
print("\nAfter insert:")
link.display()
```

# Circular Linked List 

```python
class Node:
    def __init__(self):
        self.next=None
        self.data=None
        
class circularLinkedList:
    
    def __init__(self):
        self.head=None
        
    def append(self,data):
        current=self.head
        node=Node()
        node.data=data
        if(self.head==None):   
            self.head=node
            node.next=self.head
            return
        else:
            while(current.next!=self.head):
                current=current.next
            current.next=node
            node.next=self.head
            return
        
    def display(self,i):
        current=self.head
        while(current.next!=self.head):
            print(current.data,end="->")
            current=current.next
        print(current.data,end="-->")
        
#         for no. of time list print
        for j in range(1,i):
            self.display(j-1)
        return
        
        
        
cirList=circularLinkedList()
cirList.append(1)
cirList.append(2)
cirList.append(3)
cirList.append(4)
cirList.append(5)

cirList.display(2) #pass no of time list pass
```

# Stack using Singular Linked List

```python
class Node:
    def __init__(self):
        self.data=None
        self.next=None
        
class Stack:
    def __init__(self):
        self.head=None
        
    def append(self,data):
        node= Node()
        node.data=data
        if(self.head==None):
            self.head=node
        else:
            current=self.head
            while(current.next!=None):
                current=current.next
            current.next=node
            return
        
    def pop(self):
        current=self.head
        while(current.next.next!=None):
            current=current.next
        data=current.next.data
        current.next=None
#         return data
        print("\nData poped is:",data)
        return data
    
    def display(self):
        current=self.head
        if(current==None):
            print("No elements")
        else:
            while(current.next!=None):
                print(current.data,end='->')
                current=current.next
            print(current.data)
            return
            
if __name__== "__main__":
    stack=Stack()
    
    stack.append(10)
    stack.append(20)
    stack.append(30)
    stack.append(40)
    stack.append(50)
    
    stack.display()
    
    stack.pop()
    stack.pop()
    stack.display()
    
    a=stack.pop()
    print("a is",a)
    
    stack.append(90)
    stack.display()
```
